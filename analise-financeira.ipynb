```python
import numpy as np
margemDeLucro1 = {}

# Dados fornecidos
clientes = np.array(['Empresa A', 'Empresa B', 'Empresa C', 'Empresa D', 'Empresa E'])
# Faturamento Anual em R$
faturamento = np.array([150000, 230000, 120000, 180000, 210000])
# Despesa Anual em R$
despesas = np.array([100000, 150000, 90000, 140000, 170000])


quantidadeDeEmpresas = len(clientes)
print(f"Quantidade de empresas: {quantidadeDeEmpresas}")

maiorLucro = 0.0
empresaMaiorLucro = ""
```

```python
# 1- Calcular a margem de lucro

margemDeLucro = (faturamento - despesas) / faturamento

maiorMArgemDeLucro = clientes[np.argmax(margemDeLucro)]

for i in range(0, len(clientes)):
  print(f"{clientes[i]}")
  margemDeLucro1[i] = (faturamento[i]-despesas[i]) / faturamento[i]
  print(f"Margem de lucro: {margemDeLucro1[i]*100:.2f}%\n")

  if margemDeLucro1[i] > maiorLucro:
    maiorLucro = margemDeLucro1[i]
    empresaMaiorLucro = clientes[i]

print(f"O maior lucro é de {maiorLucro*100:.2f}%\nPertence à {empresaMaiorLucro}")
```

```python
# 2 - Transformação logaritmica

transformaLucro = np.log(faturamento)
transformaDespesas = np.log(despesas)

rentabilidadeLogaritmica =  transformaLucro-transformaDespesas

for i in range(quantidadeDeEmpresas):
  print(f"Rentabilidade Logaritmica da {clientes[i]}\n Valor: {rentabilidadeLogaritmica[i]:.2f}\n")

valorDeMaiorRenta = rentabilidadeLogaritmica[np.argmax(rentabilidadeLogaritmica)]
empresaDeMaiorRenta = clientes[np.argmax(rentabilidadeLogaritmica)]

print(f"Maior rentabilidade é de: {valorDeMaiorRenta:.2f} para a {empresaDeMaiorRenta}")
```

```python
# 3 - Concatenação e Salvamento

dados = np.hstack((clientes.reshape(-1,1), faturamento.reshape(-1,1), despesas.reshape(-1,1), margemDeLucro.reshape(-1,1)))

np.savetxt('dados.csv', dados, delimiter=',', header='Clientes,Lucros,Despesas,Margem de Lucro', fmt='%s')

print("Dados salvos")
```
